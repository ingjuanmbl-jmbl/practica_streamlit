```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500

df = pd.DataFrame({
    "FECHA": pd.date_range("2023-01-02", periods = n, freq = "D").to_series().sample(n, replace=True).values,
    "REGION": np.random.choice(["NORTE","SUR","CENTRO","ORIENTE"], n),
    "VENDEDOR": np.random.choice(["ANA","JUAN CARLOS","PEDRO ALBERTO","MARIA PAZ"], n),
    "PRODUCTO": np.random.choice(["LAPTOP","MONITOR LCD","TECLADO GAMER","MOUSE GAMER","AIRPODS"], n),
    "CANTIDAD_VENDIDA": np.random.randint(1,20,n),
    "PRECIO_UNITARIO": np.random.choice([1200, 350, 85, 45, 120], n),
})

df['VENTA_TOTAL'] = df['CANTIDAD_VENDIDA'] * df['PRECIO_UNITARIO']

df['FECHA'] = pd.to_datetime(df['FECHA'])
df
```

```python
df.to_csv("ventas_sint.csv", index=False)
```
